```python
def get_center_time(*args):
    '''
    function returns total and average time (time/n-1)
    spent in middle between n arm visits
    '''
    total_time = 0;
    for i in range(len(args)-1):
        try:
            entering = df_sorted.iloc[i, 6];
            leaving = df_sorted.iloc[(i+1), 5];
            temp = leaving-entering;
            total_time = total_time+temp;
        except:
            print('error')
    avg_time=total_time/(len(args)-1);
    return total_time,avg_time;        
```

```python
def check(*args):
    '''
    function checks whether n arguments are unique
    returns boolean
    '''
    sequence = [arg.lower() for arg in args]
    isunique = len(sequence)==len(set(sequence)) # set() removes repetitions
    return isunique #True for unique sequence, False if repetition
    
```

```python
def isclwise(*args):
    l = ['A', 'B', 'C', 'D', 'E', 'A', 'B', 'C', 'D']
    i = l.index(args[0])
    for arg in args:
        if l[i]==arg:
            i=i+1;
        else:
            return False;
    return True;
```

```python
def isctwise(*args):
    l = ['E', 'D', 'C', 'B', 'A', 'E', 'D', 'C', 'B']
    i = l.index(args[0])
    for arg in args:
        if l[i]==arg:
            i=i+1;
        else:
            return False;
    return True;
```

```python
def get_results(sorted_file):
    '''
    function returns list with correct and incorrect number of entries (3s, 4s, and 5s)
    list with average time for correct and incorrect entries (3s, 4s, and 5s),
    list with average center time for correct and incorrect entries (3s, 4s, and 5s)
    list with any clockwise or counterclockwise bias on correct entries (3s, 4s, and 5s)
    arguments is Detailed_ROI file previously sorted for start entry time
    '''
    correct_3s=0;
    incorrect_3s=0;
    correct_3s=0;
    correct_4s=0;
    incorrect_4s=0;
    correct_5s=0;
    incorrect_5s=0;

    clw_3s=0;
    ctw_3s=0;
    clw_4s=0;
    ctw_4s=0;
    clw_5s=0;
    ctw_5s=0;

    t_correct_3s_time=0;
    t_incorrect_3s_time=0;
    t_correct_4s_time=0;
    t_incorrect_4s_time=0;
    t_correct_5s_time=0;
    t_incorrect_5s_time=0;

    c_correct_3s_time=0;
    c_incorrect_3s_time=0;
    c_correct_4s_time=0;
    c_incorrect_4s_time=0;
    c_correct_5s_time=0;
    c_incorrect_5s_time=0;
    
    for i in range(0, total_entries-2):
        ent1 = sorted_file.iloc[i,4];
        ent2 = sorted_file.iloc[i+1,4];
        ent3 = sorted_file.iloc[i+2,4];
        if check(ent1, ent2, ent3):
            if isclwise(ent1, ent2, ent3):
                clw_3s = clw_3s+1;
            elif isctwise(ent1, ent2, ent3):
                ctw_3s=ctw_3s+1;
            correct_3s=correct_3s+1;
            t_temp, a_temp=get_sequence_time(i, i+1, i+2);
            c_temp, ac_temp=get_center_time(i, i+1, i+2);
            t_correct_3s_time=t_correct_3s_time+t_temp;
            c_correct_3s_time=c_correct_3s_time+c_temp;
        else:
            incorrect_3s=incorrect_3s+1;
            t_temp, a_temp=get_sequence_time(i, i+1, i+2);
            c_temp, ac_temp=get_center_time(i, i+1, i+2);
            t_incorrect_3s_time=t_incorrect_3s_time+t_temp;
            c_incorrect_3s_time=c_incorrect_3s_time+c_temp;
            
        try:
            ent4 = sorted_file.iloc[i+3,4]
            if check(ent1, ent2, ent3, ent4):
                if isclwise(ent1, ent2, ent3, ent4):
                    clw_4s = clw_4s+1;
                elif isctwise(ent1, ent2, ent3, ent4):
                    ctw_4s=ctw_4s+1;
                correct_4s=correct_4s+1;
                t_temp, a_temp=get_sequence_time(i, i+1, i+2, i+3);
                c_temp, a_temp=get_center_time(i, i+1, i+2, i+3);
                t_correct_4s_time=t_correct_4s_time+t_temp;
                c_correct_4s_time=c_correct_4s_time+c_temp;
            else:
                incorrect_4s=incorrect_4s+1;
                t_temp, a_temp=get_sequence_time(i, i+1, i+2, i+3);
                c_temp, a_temp=get_center_time(i, i+1, i+2, i+3);
                t_incorrect_4s_time=t_incorrect_4s_time+t_temp;
                c_incorrect_4s_time=c_incorrect_4s_time+c_temp;
        except:
            print('no more 4s');
        try:
            ent5 = sorted_file.iloc[i+4,4];
            if check(ent1, ent2, ent3, ent4, ent5):
                if isclwise(ent1, ent2, ent3, ent5):
                    clw_5s = clw_5s+1;
                elif isctwise(ent1, ent2, ent3, ent4, ent5):
                    ctw_5s=ctw_5s+1;
                correct_5s=correct_5s+1;
                t_temp, a_temp=get_sequence_time(i, i+1, i+2, i+3, i+4);
                c_temp, a_temp=get_center_time(i, i+1, i+2, i+3, i+4);
                t_correct_5s_time=t_correct_5s_time+t_temp;
                c_correct_5s_time=c_correct_5s_time+c_temp;
    
            else:
                incorrect_5s=incorrect_5s+1;
                t_temp, a_temp=get_sequence_time(i, i+1, i+2, i+3, i+4);
                c_temp, a_temp=get_center_time(i, i+1, i+2, i+3, i+4);
                t_incorrect_5s_time=t_incorrect_5s_time+t_temp;
                c_incorrect_5s_time=c_incorrect_5s_time+c_temp;
        except:
            print('no more 5s');
       
        avg_correct_3s_time=correct_3s and t_correct_3s_time/correct_3s or 0;
        avg_incorrect_3s_time=incorrect_3s and t_incorrect_3s_time/incorrect_3s or 0;
        avg_correct_4s_time=correct_4s and t_correct_4s_time/correct_4s or 0;
        avg_incorrect_4s_time=incorrect_4s and t_incorrect_4s_time/incorrect_4s or 0;
        avg_correct_5s_time=correct_5s and t_correct_5s_time/correct_5s or 0;
        avg_incorrect_5s_time=incorrect_5s and t_incorrect_5s_time/incorrect_5s or 0;

        avg_c_correct_3s_time = correct_3s and c_correct_3s_time/correct_3s or 0;
        avg_c_incorrect_3s_time = incorrect_3s and c_incorrect_3s_time/incorrect_3s or 0;
        avg_c_correct_4s_time = correct_4s and c_correct_4s_time/correct_4s or 0;
        avg_c_incorrect_4s_time = incorrect_4s and c_incorrect_4s_time/incorrect_4s or 0;
        avg_c_correct_5s_time = correct_5s and c_correct_5s_time/correct_5s or 0;
        avg_c_incorrect_5s_time = incorrect_5s and c_incorrect_5s_time/incorrect_5s or 0;  
    
    return([correct_3s, incorrect_3s, correct_4s, incorrect_4s, correct_5s, incorrect_5s],
           [avg_correct_3s_time, avg_incorrect_3s_time, avg_correct_4s_time, avg_incorrect_4s_time,
            avg_correct_5s_time, avg_incorrect_5s_time], [clw_3s, ctw_3s, clw_4s, ctw_4s, clw_5s, ctw_5s],
          [avg_c_correct_3s_time, avg_c_incorrect_3s_time, avg_c_correct_4s_time, avg_c_incorrect_4s_time,
           avg_c_correct_5s_time, avg_c_incorrect_5s_time]);            
```

```python
def get_performances(score_list, total_entries):
    performance_fig, performance_ax = plt.subplots()
    performance_ax.set_ylim(0,1)
    performance_ax.set_ylabel("correct alternations (%)")
    performance_ax.set_title("Spontaneous Alternation Performance")
    performance_ax.set_xmargin(0.25)
    
    p3s = score_list[0]/(total_entries-2)
    p4s = score_list[2]/(total_entries-3)
    p5s = score_list[4]/(total_entries-4)

    print(p3s)
    print(p4s)
    print(p5s)

    performance_ax.plot('3s', p3s, 'r_', ms=50, mew=5)
    performance_ax.plot('4s', p4s, 'r_', ms=50, mew=5)
    performance_ax.plot('5s', p5s, 'r_', ms=50, mew=5)
    
    return performance_fig;
```

```python
def get_seq_time(seq_time_list):
    '''
    funcion to plot individual comparisons between time for correct vs incorrect for x arm entries
    argument is a list with avg of all time groups
    -need to implement for avereging multiple animals
    '''
    c3s = seq_time_list[0]
    i3s = seq_time_list[1]
    c4s = seq_time_list[2]
    i4s = seq_time_list[3]
    c5s = seq_time_list[4]
    i5s = seq_time_list[5]
    t_fig_3s, ax_3s = plt.subplots()
    ax_3s.set_ylabel("time (s)")
    ax_3s.set_title("Time for completing 3s")
    ax_3s.set_xmargin(0.5)    
    ax_3s.set_ylim(0,50)
    ax_3s.plot('correct 3s', c3s, 'k_', ms=50, mew=5)
    ax_3s.plot('incorrect 3s', i3s, 'k_', ms=50, mew=5)
    
    t_fig_4s, ax_4s = plt.subplots()
    ax_4s.set_ylabel("time (s)")
    ax_4s.set_title("Time for completing 4s")
    ax_4s.set_xmargin(0.5) 
    ax_4s.set_ylim(0,50)
    ax_4s.plot('correct 4s', c4s, 'k_', ms=50, mew=5)
    ax_4s.plot('incorrect 4s', i4s, 'k_', ms=50, mew=5)

    t_fig_5s, ax_5s = plt.subplots()
    ax_5s.set_ylabel("time (s)")
    ax_5s.set_title("Time for completing 5s")
    ax_5s.set_xmargin(0.5) 
    ax_5s.set_ylim(0,50)
    ax_5s.plot('correct 5s', c5s, 'k_', ms=50, mew=5)
    ax_5s.plot('incorrect 5s', i5s, 'k_', ms=50, mew=5)
    
    return t_fig_3s, t_fig_4s, t_fig_5s;
```

```python
def get_bias(clw_ctw_list, score_list):
    '''
    funcion to plot whether there is clockwise or conterclockwise bias
    counter-clockwise alternation subtracted from the number of clockwise, 
    divided by the total number of correct alternations.
    '''
    bias_fig, bias_ax = plt.subplots();
    bias_ax.set_ylabel("bias");
    bias_ax.set_title("Clockwise or Counterclockwise bias");
    bias_ax.set_xmargin(0.25);
    bias_ax.set_ylim(-1,1);
    
    clw3s = clw_ctw_list[0];
    ctw3s = clw_ctw_list[1];
    c3s = score_list[0];
    clw4s = clw_ctw_list[2];
    ctw4s = clw_ctw_list[3];
    c4s = score_list[2];
    clw5s = clw_ctw_list[4];
    ctw5s = clw_ctw_list[5];
    c5s = score_list[4];

    b3s = c3s and clw3s-ctw3s/c3s or 0;
    b4s = c4s and clw4s-ctw4s/c4s or 0;
    b5s = c5s and clw5s-ctw5s/c5s or 0;

    bias_ax.plot('3s', b3s, 'ko', ms=7, mew=7);
    bias_ax.plot('4s', b4s, 'ko', ms=7, mew=7);
    bias_ax.plot('5s', b5s, 'ko', ms=7, mew=7);
    
    return bias_fig;
```

```python
'''
funtion to return the beginning of exit for directionality purposes
'''
def get_directionality_exit_frame(new_df_boolean, df_direct_boolean, entry_end, entry_start, letter):
    d = 0;
    if letter=='A':
        col_i = 11;
    elif letter=='B':
        col_i = 12;
    elif letter=='C':
        col_i = 13;
    elif letter=='D':
        col_i = 14;
    elif letter=='E':
        col_i = 15;
    frames = new_df_boolean.iloc[:,col_i]
    for f in frames[entry_end:entry_start:-1]:
        if df_direct_boolean[f]==1:
            d = f+1;
            break;
    return d;
```

```python
'''
function to write df_ordered_direct
start is the same as new_df_ordered but end of entry is end for df_direct (different ROIs)
using list new_df_ordered to match with raw df_direct
generate new file new_df_direct (same entry order but different end times)
'''
def get_directionality_list(new_df_ordered, new_df_boolean, df_direct_boolean, mouse_id):
    new_df_direct = new_df_ordered;
    total_entries = len(new_df_ordered)
    for ent in total_entries:
        #get end frame
        letter = new_df_ordered.iloc[ent,0] #columm with arm entry
        ent_start = new_df_ordered.iloc[ent, 3] #column with start frame num
        ent_end = new_df_ordered.iloc[ent, 4]; #column with end frame num
        #funtion to go back from end frame until boolean in df_direct_boolean is 1 again
        new_ent_end = get_directionality_exit_frame(new_df_boolean, df_direct_boolean, ent_start, ent_end, letter);
        # edit entry end frame, time, and duration in file
        new_df_direct.iloc[ent,4] = np.int64(new_ent_end); # change entry end frame
        new_df_direct.iloc[ent,5] = np.int64((new_ent_end+1)*1/30); # change entry end time
        new_df_direct.iloc[ent,6] = new_df_direct.iloc[ent,5]-new_df_direct.iloc[ent,2] # change durationb
    new_df_direct.to_csv(mouse_id+'_direct_entries.csv');
    return new_df_direct;
```

```python
'''
function to get frames in between and give back a list of coefficients for each arm
'''
def get_coeff(df_features, start, end):
    a_count=0;
    b_count=0;
    c_count=0;
    d_count=0;
    e_count=0;
    none_count=0;
    for i in range(start, end+1):
        none=0
        if df_features.iloc[i, a column num]:
            a_count=a_count+1;
        else:
            none=none+1;
        if df_features.iloc[i, b column num]:
            b_count=b_count+1;
        else:
            none=none+1;
        if df_features.iloc[i, c column num]:
            c_count=c_count+1;
        else:
            none=none+1;        
        if df_features.iloc[i, d column num]:
            d_count=d_count+1;
        else:
            none=none+1;            
        if df_features.iloc[i, e column num]:
            e_count=e_count+1;
        else:
            none=none+1;
        if none==5:
            none_count=none_count+1;
    tot = a_count+b_count+c_count+d_count+e_count+none_count;
    a_coeff = a_count/tot;
    b_coeff = b_count/tot;
    c_coeff = c_count/tot;
    d_coeff = d_count/tot;
    e_coeff = a_count/tot;
    return a_coeff, b_coeff, c_coeff, d_coeff, e_coeff;
```

```python
'''
function to get valid entry, nose frame and return true frame entry (tail base entry) 
entry latency (nose entry to tail base entry),
true frame exit (tail base exit)
exit latency (nose exit to tail base exit)
and calculated entry duration
'''
def get_true_entry(nose_start, nose_end, c, new_df_boolean):
    frames = new_df_boolean.iloc[:,c]
    i1 = nose_start;
    ts = 0;
    while frames[i1]==0:
        ts = i1;
        i1+=1;
    true_start = ts+1;
    i2 = true_start;
    te=0;
    while frames[i2]==1:
        te = i2;
        i2+=1;
    true_end = te;
    #for time difference calculations add 1 for ending
    entry_latency =  (true_start - nose_start+1)/30;
    exit_latency = (true_end - nose_end+1)/30;
    duration = (true_end-true_start+1)/30;
    return true_start, true_end, entry_latency, exit_latency, duration;
```
